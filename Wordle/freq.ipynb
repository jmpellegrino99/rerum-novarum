```python
import sqlite3

conn = sqlite3.connect("wordle.db")
df.to_sql("freq",conn,if_exists="replace")
```

```python
query = """
select
substr(word,1,1) first_letter
, avg(match_rate) avg_match_rate
, max(match_rate) max_match_rate
from freq
group by 1
order by 2 desc
;
"""

pd.read_sql_query(query,conn)
```
