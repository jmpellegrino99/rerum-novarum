```python
import folium

# Create a map centered on the United States
m = folium.Map(location=[39.8333, -98.5855], zoom_start=4)

green_square_icon = folium.DivIcon(
    icon_size=(20, 20)  # Adjust the size of the square
    , html='<div style="width: 100%; height: 100%; background-color: lightgreen;"></div>'
)

yellow_square_icon = folium.DivIcon(
icon_size=(20, 20)  # Adjust the size of the square
, html='<div style="width: 100%; height: 100%; background-color: yellow;"></div>'
)

# Add a marker for New York City
folium.Marker(
    [40.7128, -74.0060]
    , icon=green_square_icon
    , popup='Urbs Nova York').add_to(m)

# Add a marker for Titanic
folium.Marker(
    [41.726931, -49.948253]
    , icon=yellow_square_icon
    , popup='Titanic').add_to(m)

# Save the map as an HTML file
m
```
